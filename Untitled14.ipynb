Regression problems: These are problems where our predictions have continous value attribute.(for example: Housing price prediction given different factors as input in the dataset, here house prices are continuous value numbers)

Classification problems: These problems have predictions that are discrete valued (for example: output is there 0(or)1 for binary classification)

Types of problems in Unsupervised learning:

Given, a dataset the unsupervised learning algorithms will try to find some structure of pattern in the data. It decides that data lives in separate clusters. An example of this is "Clustering Algorithm".

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate simple data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise

# Add a column of ones to X for the intercept (bias term)
m = len(X)
X_new = np.ones((m, 2))
X_new[:, 1] = X[:, 0]

# Initialize parameters (theta0 and theta1)
theta = np.random.randn(2, 1)

# Hyperparameters
learning_rate = 0.1
epochs = 100

# Gradient descent
for epoch in range(epochs):
    predictions = X_new.dot(theta)
    error = predictions - y
    gradients = (2/m) * X_new.T.dot(error)
    theta = theta - learning_rate * gradients

print("Learned parameters from scratch:")
print("Theta0 (intercept):", theta[0][0])
print("Theta1 (slope):", theta[1][0])

# Plot the result
plt.scatter(X, y, label='Data')
plt.plot(X, X_new.dot(theta), color='red', label='Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# ======================
# DATA GENERATION
# ======================
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 samples, 1 feature
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise

# ======================
# LINEAR REGRESSION - NUMPY
# ======================
print("\n" + "="*10 + " NUMPY IMPLEMENTATION " + "="*10)

def linear_regression_numpy(X, y, lr=0.1, epochs=1000):
    m = len(X)

    # Add column of 1s for bias term
    X_new = np.ones((m, 2))
    X_new[:, 1] = X[:, 0]

    # Initialize weights randomly (theta0 and theta1)
    theta = np.random.randn(2, 1)

    # Gradient descent
    for epoch in range(epochs):
        predictions = X_new.dot(theta)
        errors = predictions - y
        gradients = (2 / m) * X_new.T.dot(errors)
        theta = theta - lr * gradients

    return theta

theta_numpy = linear_regression_numpy(X, y)
print("Weights (theta) from NumPy:")
print("Intercept:", theta_numpy[0][0])
print("Slope:", theta_numpy[1][0])

# ======================
# LINEAR REGRESSION - PYTORCH
# ======================
print("\n" + "="*10 + " PYTORCH IMPLEMENTATION " + "="*10)

# Convert to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Define a basic linear regression model
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)  # 1 input  1 output

    def forward(self, x):
        return self.linear(x)

model = LinearRegressionModel()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

# Training loop
for epoch in range(1000):
    y_pred = model(X_tensor)
    loss = criterion(y_pred, y_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Extract weights
with torch.no_grad():
    weight = model.linear.weight.item()
    bias = model.linear.bias.item()

print("Weights from PyTorch:")
print("Slope (weight):", weight)
print("Intercept (bias):", bias)

# ======================
# PLOT RESULTS
# ======================
# NumPy Prediction
X_plot = np.linspace(0, 2, 100).reshape(-1, 1)
X_plot_b = np.c_[np.ones((100, 1)), X_plot]
y_pred_numpy = X_plot_b.dot(theta_numpy)

# PyTorch Prediction
X_plot_tensor = torch.tensor(X_plot, dtype=torch.float32)
with torch.no_grad():
    y_pred_torch = model(X_plot_tensor).numpy()

# Plotting
plt.scatter(X, y, label='Data', alpha=0.6)
plt.plot(X_plot, y_pred_numpy, label='NumPy Prediction', color='red')
plt.plot(X_plot, y_pred_torch, label='PyTorch Prediction', color='green', linestyle='--')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression (NumPy & PyTorch)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate simple binary classification data
np.random.seed(42)
X = np.random.randn(100, 2)  # 100 points in 2D
y = (X[:, 0] + X[:, 1] > 0).astype(int).reshape(-1, 1)  # Class 1 if x + y > 0

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic regression training
def logistic_regression(X, y, lr=0.1, epochs=1000):
    m = len(X)           # number of examples
    n = X.shape[1]       # number of features

    # Add bias column (1s)
    X_new = np.ones((m, n + 1))
    X_new[:, 1:] = X

    # Initialize weights
    theta = np.random.randn(n + 1, 1)

    # Gradient descent
    for epoch in range(epochs):
        z = X_new.dot(theta)
        h = sigmoid(z)
        gradient = (1 / m) * X_new.T.dot(h - y)
        theta = theta - lr * gradient

    return theta

theta = logistic_regression(X, y)
print("Learned weights (from scratch):")
print(theta)
```

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# ======================
# DATA GENERATION
# ======================
np.random.seed(42)
X = np.random.randn(100, 2)  # 100 samples, 2 features
y = (X[:, 0] + X[:, 1] > 0).astype(int).reshape(-1, 1)  # Binary label: 1 if x + y > 0

# ======================
# LOGISTIC REGRESSION - NUMPY
# ======================
print("\n" + "="*10 + " NUMPY IMPLEMENTATION " + "="*10)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logistic_regression_numpy(X, y, lr=0.1, epochs=1000):
    m = len(X)
    n = X.shape[1]

    # Add bias term (column of 1s)
    X_new = np.ones((m, n + 1))
    X_new[:, 1:] = X

    # Initialize weights
    theta = np.random.randn(n + 1, 1)

    # Gradient descent
    for epoch in range(epochs):
        z = X_new.dot(theta)
        h = sigmoid(z)
        gradient = (1 / m) * X_new.T.dot(h - y)
        theta -= lr * gradient

    return theta

theta_numpy = logistic_regression_numpy(X, y)
print("Weights (theta) from NumPy:")
print(theta_numpy)

# ======================
# LOGISTIC REGRESSION - PYTORCH
# ======================
print("\n" + "="*10 + " PYTORCH IMPLEMENTATION " + "="*10)

# Convert to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Define simple logistic regression model
class LogisticRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 1)  # 2 input features  1 output

    def forward(self, x):
        return torch.sigmoid(self.linear(x))

# Model, loss, and optimizer
model = LogisticRegressionModel()
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

# Training loop
for epoch in range(1000):
    y_pred = model(X_tensor)
    loss = criterion(y_pred, y_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Extract weights
with torch.no_grad():
    weights = model.linear.weight.data.numpy()
    bias = model.linear.bias.data.numpy()

print("Weights from PyTorch:")
print("Weights:", weights)
print("Bias:", bias)
```
