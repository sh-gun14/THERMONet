# 🧠 Introduction to Graph Neural Networks (GNNs) – For Beginners

---

## 🔹 What is a Graph?

A **graph** is a data structure made of:

* **Nodes (vertices)**: represent entities (e.g., atoms)
* **Edges**: represent relationships or connections between nodes (e.g., bonds)

In molecules:

* Nodes = Atoms (H, C, O, N...)
* Edges = Bonds (single, double...)

```text
Example: Methane (CH₄)

       H
       |
   H - C - H
       |
       H
```

This structure can be represented as a **graph**!

---

## 🔹 What is a Graph Neural Network (GNN)?

A **GNN** is a deep learning model that learns from **graph-structured data**.

### Core idea:

> “Each node updates its representation by **aggregating information** from its neighbors.”

This is called **Message Passing**.

---

## 🔁 GNN Layer (Simplified)

Each GNN layer works like this:

```python
for each node:
    gather neighbor features
    aggregate (sum / mean / max)
    update node representation
```

So an atom learns not only from itself, but also from atoms it's bonded to.

---

## 🔹 Why GNNs for Molecules?

Molecules are **naturally graphs**, so GNNs are ideal:

* Atoms = nodes
* Bonds = edges
* Atomic number, hybridization, etc. = node features

GNNs allow us to:
✅ Predict molecular properties like energy, polarity, boiling point, etc.
✅ Handle molecules of variable size
✅ Respect the structure of chemistry (bonding, locality)

---

## 🛠️ Common GNN Components

| Component          | Description                         |
| ------------------ | ----------------------------------- |
| `GCNConv`          | Basic Graph Convolution Layer       |
| `GATConv`          | Graph Attention Layer               |
| `global_mean_pool` | Pooling: get graph-level output     |
| `DataLoader`       | Batches graphs for training         |
| `x`                | Node features (e.g., atom types)    |
| `edge_index`       | Connectivity info (from node i → j) |
| `batch`            | Info to handle multiple graphs      |

---

## 🎓 Mini GNN Pipeline for Molecules

1. **Load Molecules as Graphs**

   * Each molecule is parsed into atoms, bonds, and features

2. **Pass Through GNN**

   * Layers like `GCNConv` apply message passing

3. **Pool Features**

   * Get 1 vector per molecule (via `global_mean_pool`)

4. **Predict Property**

   * Final linear layer predicts energy, polarity, etc.

---

## 🎨 Visualization of GNN Flow

```text
Atom Graph → GCN Layer 1 → GCN Layer 2 → Global Pool → Linear → Property
```

---

## 📚 Useful Terms to Teach

| Term            | Meaning                                      |
| --------------- | -------------------------------------------- |
| Message Passing | Nodes exchange info with neighbors           |
| Aggregation     | Combine neighbor messages (mean, sum, etc.)  |
| Pooling         | Convert node info → graph info               |
| Node Features   | What each atom "knows" (atomic number, etc.) |
| Edge Index      | Connectivity (who's bonded to whom)          |

---

## 🧪 Example: Predicting Dipole Moment (`mu`)

Given a molecule:

* Input: atom types, bond structure
* GNN processes it
* Outputs: predicted dipole moment

This is **regression on graphs**.

---

## 🧠 Summary

| Concept           | Purpose                      |
| ----------------- | ---------------------------- |
| GNN               | Learns from graph data       |
| GCN Layer         | Aggregates neighbor info     |
| Molecule as Graph | Atoms → nodes, bonds → edges |
| Application       | Predict molecular properties |

````python
import os
import kagglehub

# Download the dataset
path = kagglehub.dataset_download("zaharch/quantum-machine-9-aka-qm9")
print("📁 Downloaded to:", path)


````

````python
from torch_geometric.datasets import QM9

# Without transform
dataset = QM9(root='qm9_data')  # ✅ Works in all versions

print(dataset[0])
print(dataset[0].y)
````

````python
from torch_geometric.loader import DataLoader
import torch
from torch.nn import Linear
from torch_geometric.nn import GCNConv, global_mean_pool
import torch.nn.functional as F

# Use subset for quick testing
train_dataset = dataset[:100000]
test_dataset = dataset[100000:]
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)

class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = GCNConv(dataset.num_node_features, 64)
        self.conv2 = GCNConv(64, 64)
        self.lin = Linear(64, 1)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

# Training setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GCN().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = torch.nn.L1Loss()
target_idx = 4  # e.g., mu

# Training loop
for epoch in range(1, 6):
    model.train()
    for batch in train_loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        out = model(batch)
        loss = loss_fn(out, batch.y[:, target_idx:target_idx+1])
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
````

````python
from sklearn.metrics import mean_absolute_error

model.eval()
preds = []
trues = []

with torch.no_grad():
    for batch in test_loader:
        batch = batch.to(device)
        out = model(batch).cpu().squeeze()
        y_true = batch.y[:, target_idx].cpu().squeeze()
        preds.append(out)
        trues.append(y_true)

preds = torch.cat(preds)
trues = torch.cat(trues)

mae = mean_absolute_error(trues.numpy(), preds.numpy())
print(f"🧪 Test MAE: {mae:.4f}")
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(trues.numpy(), preds.numpy(), alpha=0.3)
plt.xlabel("True mu")
plt.ylabel("Predicted mu")
plt.title("Prediction vs. Ground Truth")
plt.plot([trues.min(), trues.max()], [trues.min(), trues.max()], 'r--')
plt.grid(True)
plt.show()
````
